```python

#Create Data frame
import pandas as pd
data={'Name':['A' ,'B' ,'C'],
      'Marks':[85,90,78]}

df=pd.DataFrame(data)
display(df)

#dimention
display(df.ndim)

#size
display(df.size)

#shape
display(df.shape) #will tell no of row and coloumn

#display coloumn names
display(df.columns)

#Check data types of each coloumn
display(df.dtypes)

#Acess Individual Coloumn
display(df['Name'])

#acccess specific Row
display(df.loc[1])

#specific no of row and coloum , direct indexing
display(df.iloc[1,1])

#Add a new coloumn
df["Grade"] = ["First class","Distinction","second class"]
display(df)

#Update Data in Data Frame
#change marks of student A to 88.
df.loc[0,"Marks"]=88
display(df)

#Delete a coloumn
#Assigning to df1 to fix the NameError
df1 = df.drop("Grade",axis=1)
display(df1)

#Basic Satastical Analysis
#mean
mean1 = df["Marks"].mean()
display(mean1)

#Min
min1 = df["Marks"].min()
display(min1)

#max
max1 = df["Marks"].max()
display(max1)

#apply condition(filtering)
#DisplAY STUDENTS SCORING MORE THAN 80
display(df[df["Marks"]>80])
```
